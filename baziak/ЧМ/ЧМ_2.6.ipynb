```python
from math import *
from matplotlib.pyplot import *
```

```python
def f1(x):
    # функция f(x) 
    return log10(x) - cos(x)

x = [0.1 + i * (4 - 0.1) / 499 for i in range(500)]
y = [f1(xi) for xi in x]

figure(1)
plot(x, y)
grid(True)
axhline(y=0, color='r')
title('lg x - cos x = 0')
xlabel('x')
ylabel('f(x)')
show()

# графически корень примерно между 1 и 2
a, b = 1, 2
eps = 0.00005  # для 4 знаков после запятой
n = 0

while b - a > eps: # циикл с условием пока мы не достигнем точности
    n += 1
    print(f"Итерация: {n} - Корень: {(a + b) / 2:.5f}")
    c = (a + b) / 2
    if f1(a) * f1(c) <= 0:
        b = c
    else:
        a = c

root1 = (a + b) / 2
print(f"Итерация с подходящей точностью: {n}")
print(f"Корень: {root1:.5f}")
```

Упражнение 2

```python
def f2(x):
    # функция где надл найти корни
    return x**2 - 3.2*x - 1

# создаем точки для графика вручную без numpy
x_vals2 = [-2 + i * (5 - (-2)) / 1000 for i in range(1001)]
y_vals2 = [f2(x) for x in x_vals2]

# строим график
figure(figsize=(12, 8))
plot(x_vals2, y_vals2, 'r-', linewidth=2, label='f(x)')
axhline(y=0, color='k', linestyle='-', alpha=0.3)
axvline(x=0, color='k', linestyle='-', alpha=0.3)
grid(True, alpha=0.3)
xlabel('x')
ylabel('f(x)')
title('Графическое отделение корней для уравнения')
legend()

# точные корни через дискриминант
d = 3.2**2 + 4
root1 = (3.2 + sqrt(d)) / 2
root2 = (3.2 - sqrt(d)) / 2
print(f"Точные корни: {root1:.4f} и {root2:.4f}")

# отмечаем корни на графике
plot(root1, 0, 'bo', markersize=8, label=f'Корень 1: x ≈ {root1:.3f}')
plot(root2, 0, 'go', markersize=8, label=f'Корень 2: x ≈ {root2:.3f}')
legend()
show()
```

Упражнение 3

```python
def f3(x):
    # нужная функция
    return 4*x - 3*log(x) - 4

# итерационная функция
def g3(x):
    return (4 + 3*log(x))/4

# создаем точки для графика
x_vals = [0.1 + i * (4 - 0.1) / 499 for i in range(500)]
y_vals = [f3(x) for x in x_vals]

# находим корень методом итерации
x = 2.0  # начальное приближение
eps = 1e-4
for _ in range(100):
    x_new = g3(x)
    if abs(x_new - x) < eps:
        break
    x = x_new
root = x_new

# строим график
figure(figsize=(12, 6))
plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = 4x - 3lnx - 4')
axhline(0, color='k', linestyle='-')
axvline(0, color='gray', alpha=0.5)
grid(True, alpha=0.3)

# отмечаем корень на нашем отрезке
plot(root, 0, 'ro', markersize=8, label=f'корень: x = {root:.4f}')
plot(root, f3(root), 'ro', markersize=8)

# вывод графика и подписи
title('уравнение 4x - 3ln x = 4')
xlabel('x')
ylabel('f(x)')
legend()
xlim(0, 3)
ylim(-3, 5)
show()
# вывод корня
print(f"Корень: {root:.4f}")
```
